```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df
```

```python
df.info()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
df
```

```python
df['Geography'].dtype
```

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
df
```

```python
X=df.drop('Exited',axis=1)
y=df['Exited']
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train_scl=StandardScaler().fit_transform(X_train)
X_test_scl= StandardScaler().fit_transform(X_test)
```

```python
X_train_scl
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
cnn=Sequential()
cnn.add(Dense(11,activation='relu',kernel_initializer='he_normal'))
cnn.add(Dense(11,activation='relu'))
cnn.add(Dense(1,activation ='sigmoid'))
```

```python
cnn.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.built
```

```python
cnn.fit(X_train_scl,y_train,batch_size=10,epochs=10,verbose='auto',validation_split=0.2)
```
